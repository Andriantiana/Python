<h1>Introduction to DBSCAN algorithm </h1>
<p>DBSCAN is a clustering algorithm based on density</p>

```python
#First, import the packages
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
#Built a plot of points
X, y = make_blobs(random_state=170, n_samples=600, centers = 5)
#Stretch the plot
rng = np.random.RandomState(74)
transformation = rng.normal(size=(2, 2))
X = np.dot(X, transformation)
plt.scatter(X[:, 0], X[:, 1])
#Show the plot
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

<p>Obviously, there is 5 clusters from this data, let's see how DBSCAN manages it<p>

```python
#Apply DBCSCAN
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# cluster the data into five clusters
dbscan = DBSCAN(eps=0.123, min_samples = 2)
clusters = dbscan.fit_predict(X_scaled)
# plot the cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="plasma")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

<p>Notice there are 3 points that haven't been put on a cluster, that because there are considered as noise, it's a big difference with DBSCAN comparing to other clustering method like Kmeans </p>

```python
# cluster the data into five clusters
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
y_pred = kmeans.predict(X)
# plot the cluster assignments and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap="plasma")
plt.scatter(kmeans.cluster_centers_[:, 0],   
            kmeans.cluster_centers_[:, 1],
            marker='^', 
            c=[0, 1, 2, 3, 4], 
            s=100, 
            linewidth=2,
            cmap="plasma")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
